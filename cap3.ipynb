### Cap 3 - Introdução às listas

```python
# Nomes
amigas = ["Sâmia", "Didia", "Lorena", "Livia"]
for amiga in amigas:
    print(f"Minha amiga {amiga} é muito bacana")
```

```python
# Adicionando elementos
# Ao final
amigas.append("Aline")
print(amigas)
```

```python
# Inserindo em uma posição
amigas.insert(2, "Duda")
print(amigas)
```

```python
# Retirando elementos
del amigas[0]
print(amigas)
```

```python
# pop
amiga_retirada = amigas.pop()
print(amiga_retirada)
print(amigas)

amigas.pop(1)
print(amigas)
```

```python
# remove
amigas.remove("Didia")
print(amigas)
```

```python
# Exemplo do livro
motorcycles = ["honda", "yamaha", "suzuki", "ducati"]
print(motorcycles)
too_expansive = "ducati"
motorcycles.remove(too_expansive)
print(f"\nA {too_expansive.title()} is too expansive for me.")
```

```python
# Organizando uma lista
motorcycles.sort()
print(motorcycles)
```

```python
motorcycles.sort(reverse=True)
print(motorcycles)
```

```python
print(f"\n Here is the original list: {motorcycles}")
print(f"\n Here is a sorted list: {sorted(motorcycles)}")
```

```python
lista_numerica = [0, 1, 3, 5, 7, 9]
print(lista_numerica)
lista_numerica.reverse()
print(lista_numerica)
lista_numerica.reverse()
print(lista_numerica)
```
