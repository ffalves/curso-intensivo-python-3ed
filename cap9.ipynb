### Cap 9 - Classes

```python
class Dog:
    """Tentativa de modelar um cão."""

    def __init__(self, name, age):
        """Inicializa os atributos de nome e idade."""
        self.name = name
        self.age = age

    def sit(self):
        """Simula um cachorro sentado em resposta a um comando."""
        print(f"{self.name} is now sitting.")

    def roll_over(self):
        """Simula um cachorro rolando em resposta a um comando."""
        print(f"{self.name} rolled over!")
```

```python
my_dog = Dog("Willie", 6)
print(f"My dog's name {my_dog.name} and is {my_dog.age} years old.")
my_dog.sit()
my_dog.roll_over()
```

```python
class Restaurant:
    """Tentativa de modelar um restaurante."""

    def __init__(self, restaurant_name, cuisine_type):
        """Inicializa os atributos de nome e tipo de gastronomia."""
        self.restaurant_name = restaurant_name
        self.cuisine_type = cuisine_type

    def describe_restaurant(self):
        """Apresenta uma descrição do restaurante."""
        print(
            f"The restaurant {self.restaurant_name} is specialized in {self.cuisine_type}."
        )

    def open_restaurant(self):
        """Sinaliza se o restaurante está aberto e funcionando."""
        print(f"The {self.restaurant_name} is open now!")
```

```python
# Restaurant example
korean_restaurant = Restaurant("Portal da Coreia", "comida coreana")
korean_restaurant.describe_restaurant()
korean_restaurant.open_restaurant()
```

```python
# User example
class User:
    """Tentativa de modelar um perfil de usuário qualquer."""

    def __init__(self, first_name, last_name, age, gender, hometown):
        """Inicializa os atributos de nome e sobrenome do usuário."""
        self.first_name = first_name
        self.last_name = last_name
        self.age = age
        self.gender = gender
        self.hometown = hometown

    def describe_user(self):
        """Apresenta uma descrição do usuário"""
        print(
            f"{self.first_name} {self.last_name} is {self.age} years old and {self.gender} was born in {self.hometown}."
        )

    def greet_user(self):
        """Cumprimenta o usuário."""
        print(f"Hello, {self.first_name}, welcome!")
```

```python
my_user = User("Fefe", "Alves", 42, "she", "Teresina-PI")
my_user.describe_user()
my_user.greet_user()
```
